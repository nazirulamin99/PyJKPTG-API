```python
df = duckdb.sql("SELECT * FROM './data/Licenselist-20260122.parquet'").df()
print(f"Rows: {len(df)}")
print(f"\nUnique region: {df['region'].nunique()} -> {df['region'].unique().tolist()[:5]}...")
print(f"Unique company: {df['company'].nunique()}")
print(f"Unique shipName: {df['shipName'].nunique()}")
print(f"Unique licenseNo: {df['licenseNo'].nunique()}")
print(f"Unique concession: {df['concession'].nunique()} -> {df['concession'].unique().tolist()[:5]}")
print(f"Unique project: {df['project'].nunique()} -> {df['project'].unique().tolist()[:5]}")
print(f"Unique noFile: {df['noFile'].nunique()} -> {df['noFile'].unique().tolist()[:5]}")
print(f"\nSample row:")
print(df.iloc[0].to_dict())
```
